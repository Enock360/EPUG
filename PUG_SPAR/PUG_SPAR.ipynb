```python
import torch
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import os 
import shutil
```

## Specify the path of the dataset there

```python
PUG_SPAR_PATH="TO_CHANGE"
```

## Read the PUG: SPAR csv label file

```python
main_df = pd.read_csv(PUG_SPAR_PATH+"labels.csv")
```

## Plot random images from the dataset

```python
f, axarr = plt.subplots(8,6, figsize=(12,16))
i = 0
j = 0
for k in range(48):
    index = np.random.choice(len(main_df), 1)
    img = Image.open(os.path.join(PUG_SPAR_PATH, main_df['filename'][index[0]]))
    axarr[j,i%6].axis('off')
    axarr[j,i%6].imshow(img, aspect="auto")
    i = i + 1
    j = int(i / 6)
f.subplots_adjust(wspace=.05, hspace=.05)
plt.show()
```

## Plot only the background environments

```python
subset_background_only = main_df[main_df['character_name'] == 'blank'].reset_index(drop=True)
subset_background_only = subset_background_only[subset_background_only['character2_name'] == 'blank'].reset_index(drop=True)
subset_background_only = subset_background_only[:10]
f, axarr = plt.subplots(2,5, figsize=(12,5))
i = 0
j = 0
for k in range(10):
    img = Image.open(os.path.join(PUG_SPAR_PATH, subset_background_only['filename'][k]))
    axarr[j,i%5].axis('off')
    axarr[j,i%5].imshow(img, aspect="auto")
    i = i + 1
    j = int(i / 5)
f.subplots_adjust(wspace=.05, hspace=.05)
plt.show()
```

## Show only single animal images

```python
subset_char1_only = main_df[(main_df['character_name'] == 'blank') & (main_df['character2_name'] != 'blank')]
subset_char2_only = main_df[(main_df['character2_name'] == 'blank') & (main_df['character_name'] != 'blank')]
subset_char_only = pd.concat([subset_char1_only, subset_char2_only]).reset_index(drop=True)
subset_char_only = subset_char_only[(subset_char_only['character_texture'] == 'Default')].reset_index(drop=True)
f, axarr = plt.subplots(2,5, figsize=(12,5))
i = 0
j = 0
for k in range(10):
    index = np.random.choice(len(subset_char_only), 1)
    img = Image.open(os.path.join(PUG_SPAR_PATH, subset_char_only['filename'][index[0]]))
    axarr[j,i%5].axis('off')
    axarr[j,i%5].imshow(img, aspect="auto")
    i = i + 1
    j = int(i / 5)
f.subplots_adjust(wspace=.05, hspace=.05)
plt.show()
```
